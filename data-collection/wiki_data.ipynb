```python
import requests
import pandas as pd
from bs4 import BeautifulSoup

def parse_word_page(url):
    response = requests.get(url)
    soup = BeautifulSoup(response.text, "html.parser")
    return soup
```

```python
soup = parse_word_page('https://en.wikipedia.org/wiki/Glossary_of_Generation_Z_slang')
# Find all description lists in the content
description_lists = soup.find_all("dl")

# Initialize a list to store the slang terms and definitions
slang_terms = []

# Iterate through each description list
for dl in description_lists:
    # Find all terms (<dt>) and definitions (<dd>) within the list
    terms = dl.find_all("dt")
    definitions = dl.find_all("dd")
    
    # Pair each term with its corresponding definition
    for term, definition in zip(terms, definitions):
        slang = term.get_text(strip=True)
        meaning = definition.get_text(strip=True)
        slang_terms.append((slang, meaning))

slang_df = pd.DataFrame(slang_terms, columns = ['term', 'definition'])
```
